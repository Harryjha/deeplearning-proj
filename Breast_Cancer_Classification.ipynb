Importing the Dependencies


```python
import numpy as ny
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
X = data_frame.drop(columns = 'label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,
                                                   random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Standardize. the data


```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import  keras
```

```python
# seeting up the layers of the neural network
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)), #coverting multilayer array to single layer array
                          keras.layers.Dense(20, activation='relu'), # hidden layer
                          keras.layers.Dense(2, activation='sigmoid') # output layer
])
```

```python
# compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# train the  neural network
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=15)
```

visualizing accuracy and **loss**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc = 'upper right')
```

Accuracy of the model on the test **data**

```python
loss, accuracy= model.evaluate(X_test_std, Y_test)
print(accuracy)
```
